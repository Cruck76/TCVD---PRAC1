```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = "https://www.backmarket.es/search?q=iphone"

page = requests.get(url)

#content = page.text
soup = BeautifulSoup(page.content, "html.parser")


product = soup.find_all("h2", class_="duration-200 line-clamp-1 md:mb-1 md:mt-0 mt-1 normal-case overflow-ellipsis overflow-hidden text-black transition-all font-body text-3 leading-3 font-bold")

productes = list()

for p in product:
    productes.append(p.get_text().strip())
    
print(productes)


caract = soup.find_all("span", class_="duration-200 line-clamp-1 normal-case overflow-ellipsis overflow-hidden text-black transition-all font-body text-2 leading-2 font-light")
caracteristiques = list()
for car in caract:
    caracteristiques.append(car.get_text().strip())
print(caracteristiques)


garantia = soup.find_all("span", class_="text-black font-body text-2 leading-2 font-light")
garanties = list()
for gar in garantia:
    garanties.append(gar.get_text().strip())
print(garanties)


price = soup.find_all("span", class_="text-black font-body text-2 leading-2 font-bold")
prices = list()
for titol in price:
    prices.append(titol.text.strip()[0:titol.text.find("x")-1])
print(prices)


pag = soup.find_all("a", class_="focus:outline-none group md:box-border relative w-full md:w-[26.2rem]")
pagines = list()
for titol in pag:
    pagines.append("https://www.backmarket.es"+titol.get("href"))
print(pagines)
```

```python
df = pd.DataFrame({"Producte": productes, "Característiques": caracteristiques, "Garantia": garanties, 
                   "Preu": prices, "Url": pagines})

#print(df)

# Imprimint les pagines per buscar el reacondicionador

enviat = list()
reacondicionador = list()
puntuacions = list()
    
for a in pagines:
    page = requests.get(a)

    soup = BeautifulSoup(page.content, "html.parser")

    empresa = soup.find_all("p", class_="mb-1 overflow-hidden font-body text-3 leading-3 font-bold")


    for p in empresa:
        reacondicionador.append(p.text[30:].strip())
        
    
    envio = soup.find_all("span", class_="text-grey-500 font-body text-2 leading-2 font-bold")

    for p in envio:
        enviat.append(p.text.strip())
    
    
    puntuacio = soup.find("span", class_="ml-1 text-primary font-body text-2 leading-2 font-bold")

    for p in puntuacio:
        puntuacions.append(p.text.strip()[0:titol.text.find("\/")-1])


df["Reacondicionador"] = reacondicionador
df["Origen_enviament"] = enviat
df["Puntuacio"] = puntuacions

print(df)
```

```python
df.to_csv("Prova-COMPRA-WebScraping.csv", index=False)
```
